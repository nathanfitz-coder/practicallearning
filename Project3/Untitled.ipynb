## Project 3 Vocabulary

The custom vocabulary was generated by fitting a logistic regression model to a bag of words matrix as the predictor set, and the binary review sentiment classification as the target variable. The logistic regression model yields coeficients for every word. Words with high possitive coefficients indicate words that have strong possitive sentiment such as "good" and "excellent". Conversly, words with low negative coefficents indicate words that have strong negative sentiment such as "poor" and "terible". We are interested in both sets of words. In short, we want words that have the hightest absolute value of their coefficients. Therefore, the top 1000 words with the hightest absolute value were chosen.  

The python function that generates the vocabulary list is below:

```python
def write_vocab(source='data/alldata.tsv',target='myvocab.txt'):
    source='data/alldata.tsv'
    alldata = pd.read_csv(source,sep="\t")
    
    #Clean non-letter characters, convert to lowercase, and split into individual words
    alldata['clean_reviews'] = alldata['review'].replace('[^a-zA-Z\s]','',regex=True)
    
    corpus = alldata['clean_reviews'].tolist()

    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(corpus)
    y = alldata['sentiment'].values
    #vocab = vectorizer.get_feature_names()
    
    
    # test p-values
    model = LogisticRegression(max_iter=1000).fit(X, y)
    coefs = {}
    for i, coef in enumerate(list(model.coef_[0,:])):
        coefs[i] = abs(coef)
    

    topcoefs = sorted(coefs.items(), key=lambda x: x[1], reverse=True) 
    vocab = []
    features = vectorizer.get_feature_names()
    for coef in topcoefs[0:2000]:
        vocab.append(features[coef[0]])
        
    del(features)

    with open(target, 'w') as filehandle:
        filehandle.writelines("%s\n" % word for word in vocab)
```
